```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Total day minutes'], bins=30, kde=True)
plt.title('Distribusi Total Menit Telepon Siang Hari')
plt.xlabel('Total Day Minutes')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='Customer service calls', data=df)
plt.title('Jumlah Telepon Customer Service: Churn vs Tidak Churn')
plt.xlabel('Churn (False = Bertahan, True = Berhenti)')
plt.ylabel('Jumlah Telepon ke Customer Service')
plt.show()
```

```python
pd.crosstab(df['International plan'], df['Churn'], normalize='index') * 100
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='International plan', hue='Churn', data=df)
plt.title('Jumlah Pelanggan berdasarkan International Plan dan Churn')
plt.show()
```

```python
plt.figure(figsize=(12,8))
numeric_df = df.select_dtypes(include=['float64','int64'])
correlation = numeric_df.corr()
sns.heatmap(correlation, annot=False, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap - Semua Variabel Numerik')
plt.show()
```

Insight dari EDA:
1. Dataset memiliki proporsi churn ~14.5%, tergolong imbalanced.
2. Pelanggan dengan jumlah telepon ke customer service yang tinggi (>3-4 kali) cenderung lebih banyak churn.
3. Pelanggan dengan International Plan aktif memiliki tingkat churn jauh lebih tinggi dibanding yang tidak.
4. Total day minutes dan Total day charge berkorelasi sangat kuat karena charge dihitung dari minutes.
